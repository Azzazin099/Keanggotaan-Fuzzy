Soal Ganjil

Kecepatan tendangan bola (km/jam)
1. Sangat Lambat = 10 - 20
2. Agak Lambat = 35 - 50
3. Cukup Cepat = 60 - 70
4. Cepat= 80 - 100
Toleransi/bias tentukan sendiri

```
import numpy as np #melakukan import library numpy
import matplotlib.pyplot as plt #melakukan import limbrary matplotlib untuk menampilkan grafik
import skfuzzy as sk #melakukan import library fuzzy
%matplotlib inline
```

```
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as sk
%matplotlib inline

toleransi = 6 #nilai toleransi yang diberikan, nilai yang diberikan bebas, nilai toleransi yang diambil saat ini adalah 6
#paramerternya (batas bawah,bat)
x = np.arange(0,120,1) #range dari kecepatan maximal adalah 100 tapi diambil 120 dengan kenaikannya adalah 1

sangat_lambat = sk.trapmf(x,[10-toleransi,10,20,20+toleransi]) #menggunakan fungsi trapesium, dengan x adalah value dari x, 10-toleransi adalah kaki awal, 
#10 adalah puncak pertama, 20 adalah puncak kedua dan 20+toleransi adalah nilai akhir(kaki akhir)

agak_lambat = sk.trapmf(x,[35-toleransi,35,50,50+toleransi])#menggunakan fungsi trapesium, dengan x adalah value dari x, 35-toleransi adalah kaki awal,
#35 adalah puncak pertama, 50 adalah puncak kedua dan 50+toleransi adalah nilai akhir(kaki akhir)

cukup_cepat = sk.trapmf(x,[60-toleransi,60,70,70+toleransi])#menggunakan fungsi trapesium, dengan x adalah value dari x, 60-toleransi adalah kaki awal,
#60 adalah puncak pertama, 70 adalah puncak kedua dan 70+toleransi adalah nilai akhir(kaki akhir)

cepat = sk.trapmf(x,[80-toleransi,80,100,100+toleransi]) #menggunakan fungsi trapesium, dengan x adalah value dari x, 80-toleransi adalah kaki awal,
#80 adalah puncak pertama, 100 adalah puncak kedua dan 100+toleransi adalah nilai akhir(kaki akhir)

plt.plot(x,sangat_lambat) #melakukan plot untuk variabel sangat_lambat
plt.plot(x,agak_lambat) #melakukan plot untuk variabel agak_lambat
plt.plot(x,cukup_cepat)#melakukan plot untuk variabel cukup_cepat
plt.plot(x,cepat) #melakukan plot untuk variabel  cepat

plt.legend(["Sangat Lambat", "Agak Lambat","Cukup Cepat", "Cepat",])
plt.title("Derajat Keanggotaan Cepat Bola",fontsize=12)
plt.xlabel("Kecepatan km/jam")
```
